# Interactive Business Dashboard in Streamlit

## Introduction

Business dashboards provide an interactive way to analyze sales performance and support data-driven decision-making. Instead of examining raw data, dashboards summarize key business metrics through interactive charts, filters, and Key Performance Indicators (KPIs).

In this project, an interactive business dashboard is developed using Streamlit and the Global Superstore dataset. The dashboard allows users to explore sales, profit, customer performance, and product categories through dynamic filters and visualizations. It enables business users to quickly identify trends, compare regional performance, and make informed decisions based on real-time insights.

## Problem Statement

Businesses generate large volumes of sales data every day. Analyzing this information using spreadsheets alone is time-consuming and makes it difficult to identify important trends and business opportunities.

To improve decision-making, an interactive dashboard is required that enables users to filter sales data by Region, Category, and Sub-Category while displaying important business metrics such as Total Sales, Total Profit, Top Customers, and sales performance through interactive visualizations.

## Project Objectives

The main objectives of this project are:

- Perform data cleaning and preprocessing.
- Prepare the dataset for dashboard development.
- Develop an interactive dashboard using Streamlit.
- Implement interactive filters for Region, Category, and Sub-Category.
- Display key performance indicators (KPIs) such as Total Sales, Total Profit, Total Orders, and Total Customers.
- Visualize business performance using interactive charts.
- Identify the Top 5 Customers by Sales.
- Support business decision-making through interactive data storytelling.

### Data Consistency Check

The dataset is examined for common data quality issues such as leading and trailing spaces in text-based features. Ensuring consistent text values improves filtering, grouping, and visualization accuracy within the interactive dashboard.

#### Check for Leading and Trailing Spaces

The dataset is examined for leading and trailing spaces across all text-based features. Removing unnecessary spaces ensures consistent categorical values and improves the accuracy of filtering, grouping, and dashboard visualizations.

```python
# Check for leading and trailing spaces in all text columns

text_columns = df.select_dtypes(include=["object"]).columns

space_summary = pd.DataFrame(
    [
        [column, (df[column] != df[column].str.strip()).sum()]
        for column in text_columns
    ],
    columns=["Column Name", "Rows with Extra Spaces"]
)

space_summary[space_summary["Rows with Extra Spaces"] > 0]
```

##### Observation

The data consistency check identified leading and trailing spaces only in the **Product Name** column, affecting **16 records**. All other text-based features were free from unnecessary spaces.

#### Remove Leading and Trailing Spaces

Leading and trailing spaces are removed from all text-based features to ensure consistent categorical values. This improves filtering, grouping, and visualization accuracy within the interactive dashboard.

```python
# Remove leading and trailing spaces from all text columns

text_columns = df_clean.select_dtypes(include=["object"]).columns

for column in text_columns:
    df_clean[column] = df_clean[column].str.strip()
```

```python
# Verify that no leading and trailing spaces remain

space_summary = pd.DataFrame(
    [
        [column, (df_clean[column] != df_clean[column].str.strip()).sum()]
        for column in text_columns
    ],
    columns=["Column Name", "Rows with Extra Spaces"]
)

space_summary[space_summary["Rows with Extra Spaces"] > 0]
```

##### Observation

Leading and trailing spaces were successfully removed from all text-based features. No columns contain unnecessary spaces, ensuring standardized values for accurate filtering, grouping, and dashboard analysis.

### Date Format Conversion

The **Order Date** and **Ship Date** columns are converted from text format to the datetime format. This conversion enables time-based analysis such as yearly trends, monthly sales, quarterly reports, and date filtering within the interactive Streamlit dashboard.

```python
# Convert date columns to datetime format

df_clean["Order Date"] = pd.to_datetime(
    df_clean["Order Date"],
    dayfirst=True
)

df_clean["Ship Date"] = pd.to_datetime(
    df_clean["Ship Date"],
    dayfirst=True
)
```

```python
# Verify updated data types

df_clean[["Order Date", "Ship Date"]].dtypes
```

#### Observation

The **Order Date** and **Ship Date** columns were successfully converted from string format to the **datetime** data type. This enables efficient time-based analysis, including yearly and monthly sales trends, and supports interactive date filtering in the Streamlit dashboard.

### Feature Engineering

Feature engineering is performed to create additional date-related features from the **Order Date** column.

These features simplify time-based analysis and improve the performance of the interactive Streamlit dashboard by enabling efficient filtering and visualization of yearly, quarterly, monthly, and daily sales trends.

```python
# Create date-related features

df_clean["Order Year"] = df_clean["Order Date"].dt.year

df_clean["Order Quarter"] = df_clean["Order Date"].dt.quarter

df_clean["Order Month"] = df_clean["Order Date"].dt.month

df_clean["Order Month Name"] = df_clean["Order Date"].dt.month_name()

df_clean["Order Day"] = df_clean["Order Date"].dt.day

df_clean["Order Day Name"] = df_clean["Order Date"].dt.day_name()
```

```python
# Display the newly created features

df_clean[
    [
        "Order Date",
        "Order Year",
        "Order Quarter",
        "Order Month",
        "Order Month Name",
        "Order Day",
        "Order Day Name"
    ]
].head()
```

#### Observation

Six additional date-related features were successfully created from the **Order Date** column. These features enable efficient yearly, quarterly, monthly, and daily analysis, improving the flexibility and interactivity of the Streamlit business dashboard.
