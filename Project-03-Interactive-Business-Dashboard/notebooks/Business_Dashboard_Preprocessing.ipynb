# Interactive Business Dashboard in Streamlit

## Introduction

Business dashboards provide an interactive way to analyze sales performance and support data-driven decision-making. Instead of examining raw data, dashboards summarize key business metrics through interactive charts, filters, and Key Performance Indicators (KPIs).

In this project, an interactive business dashboard is developed using Streamlit and the Global Superstore dataset. The dashboard allows users to explore sales, profit, customer performance, and product categories through dynamic filters and visualizations. It enables business users to quickly identify trends, compare regional performance, and make informed decisions based on real-time insights.

## Problem Statement

Businesses generate large volumes of sales data every day. Analyzing this information using spreadsheets alone is time-consuming and makes it difficult to identify important trends and business opportunities.

To improve decision-making, an interactive dashboard is required that enables users to filter sales data by Region, Category, and Sub-Category while displaying important business metrics such as Total Sales, Total Profit, Top Customers, and sales performance through interactive visualizations.

## Project Objectives

The main objectives of this project are:

- Perform data cleaning and preprocessing.
- Prepare the dataset for dashboard development.
- Develop an interactive dashboard using Streamlit.
- Implement interactive filters for Region, Category, and Sub-Category.
- Display key performance indicators (KPIs) such as Total Sales, Total Profit, Total Orders, and Total Customers.
- Visualize business performance using interactive charts.
- Identify the Top 5 Customers by Sales.
- Support business decision-making through interactive data storytelling.

### Data Consistency Check

The dataset is examined for common data quality issues such as leading and trailing spaces in text-based features. Ensuring consistent text values improves filtering, grouping, and visualization accuracy within the interactive dashboard.

#### Check for Leading and Trailing Spaces

The dataset is examined for leading and trailing spaces across all text-based features. Removing unnecessary spaces ensures consistent categorical values and improves the accuracy of filtering, grouping, and dashboard visualizations.

```python
# Check for leading and trailing spaces in all text columns

text_columns = df.select_dtypes(include=["object"]).columns

space_summary = pd.DataFrame(
    [
        [column, (df[column] != df[column].str.strip()).sum()]
        for column in text_columns
    ],
    columns=["Column Name", "Rows with Extra Spaces"]
)

space_summary[space_summary["Rows with Extra Spaces"] > 0]
```

##### Observation

The data consistency check identified leading and trailing spaces only in the **Product Name** column, affecting **16 records**. All other text-based features were free from unnecessary spaces.

#### Remove Leading and Trailing Spaces

Leading and trailing spaces are removed from all text-based features to ensure consistent categorical values. This improves filtering, grouping, and visualization accuracy within the interactive dashboard.

```python
# Remove leading and trailing spaces from all text columns

text_columns = df_clean.select_dtypes(include=["object"]).columns

for column in text_columns:
    df_clean[column] = df_clean[column].str.strip()
```

```python
# Verify that no leading and trailing spaces remain

space_summary = pd.DataFrame(
    [
        [column, (df_clean[column] != df_clean[column].str.strip()).sum()]
        for column in text_columns
    ],
    columns=["Column Name", "Rows with Extra Spaces"]
)

space_summary[space_summary["Rows with Extra Spaces"] > 0]
```

##### Observation

Leading and trailing spaces were successfully removed from all text-based features. No columns contain unnecessary spaces, ensuring standardized values for accurate filtering, grouping, and dashboard analysis.

### Date Format Conversion

The **Order Date** and **Ship Date** columns are converted from text format to the datetime format. This conversion enables time-based analysis such as yearly trends, monthly sales, quarterly reports, and date filtering within the interactive Streamlit dashboard.

```python
# Convert date columns to datetime format

df_clean["Order Date"] = pd.to_datetime(
    df_clean["Order Date"],
    dayfirst=True
)

df_clean["Ship Date"] = pd.to_datetime(
    df_clean["Ship Date"],
    dayfirst=True
)
```

```python
# Verify updated data types

df_clean[["Order Date", "Ship Date"]].dtypes
```

#### Observation

The **Order Date** and **Ship Date** columns were successfully converted from string format to the **datetime** data type. This enables efficient time-based analysis, including yearly and monthly sales trends, and supports interactive date filtering in the Streamlit dashboard.

### Feature Engineering

Feature engineering is performed to create additional date-related features from the **Order Date** column.

These features simplify time-based analysis and improve the performance of the interactive Streamlit dashboard by enabling efficient filtering and visualization of yearly, quarterly, monthly, and daily sales trends.

```python
# Create date-related features

df_clean["Order Year"] = df_clean["Order Date"].dt.year

df_clean["Order Quarter"] = df_clean["Order Date"].dt.quarter

df_clean["Order Month"] = df_clean["Order Date"].dt.month

df_clean["Order Month Name"] = df_clean["Order Date"].dt.month_name()

df_clean["Order Day"] = df_clean["Order Date"].dt.day

df_clean["Order Day Name"] = df_clean["Order Date"].dt.day_name()
```

```python
# Display the newly created features

df_clean[
    [
        "Order Date",
        "Order Year",
        "Order Quarter",
        "Order Month",
        "Order Month Name",
        "Order Day",
        "Order Day Name"
    ]
].head()
```

#### Observation

Six additional date-related features were successfully created from the **Order Date** column. These features enable efficient yearly, quarterly, monthly, and daily analysis, improving the flexibility and interactivity of the Streamlit business dashboard.

### Sales by Category

This analysis examines the total sales generated by each product category. It helps identify the highest revenue-generating categories and provides valuable insights for inventory planning, product strategy, and business decision-making.

```python
# Original dataframe (numeric values)
category_sales = (
    df_clean
    .groupby("Category", as_index=False)["Sales"]
    .sum()
    .sort_values(by="Sales", ascending=False)
)

# Display dataframe (formatted values)
category_sales_display = category_sales.copy()

category_sales_display["Sales"] = category_sales_display["Sales"].map(
    lambda x: f"${x:,.2f}"
)

category_sales_display
```

#### Sales by Category Visualization

The following bar chart visualizes the total sales generated by each product category. It provides a clear comparison of category-wise sales performance and helps identify the categories contributing the most to overall business revenue.

```python
# Create sales by category bar chart

plt.figure(figsize=(8, 5))

ax = sns.barplot(
    data=category_sales,
    x="Category",
    y="Sales"
)

# Add sales labels on bars

for container in ax.containers:
    ax.bar_label(
        container,
        fmt="%.0f",
        padding=3
    )

plt.ylim(0,5100000)
plt.ticklabel_format(style='plain', axis='y')

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales (USD)")

plt.tight_layout()

# plt.savefig(
#     "../outputs/figures/01_sales_by_category.png",
#     dpi=300,
#     bbox_inches="tight"
# )

plt.show()
```

#### Key Business Insights

- Technology generated the highest total sales, contributing approximately **4.74 million USD** to the overall business revenue.

- Furniture ranked second with total sales of approximately **4.11 million USD**, indicating strong customer demand.

- Office Supplies recorded the lowest sales among the three categories, with approximately **3.79 million USD** in total revenue.

- Technology and Furniture together contributed the majority of total sales, making them the company's strongest-performing product categories.

- All three product categories generated substantial sales, indicating a well-balanced product portfolio across the business.

### Monthly Sales Trend

This analysis examines monthly sales trends over time to identify seasonal patterns, sales fluctuations, and overall business performance. Understanding monthly sales trends helps businesses evaluate growth, forecast future demand, and make informed strategic decisions.

```python
# Calculate monthly sales

monthly_sales = (
    df_clean
    .groupby(
        ["Order Year", "Order Month", "Order Month Name"],
        as_index=False
    )["Sales"]
    .sum()
    .sort_values(["Order Year", "Order Month"])
)

# Create display dataframe

monthly_sales_display = monthly_sales.copy()

monthly_sales_display["Month"] = (
    monthly_sales_display["Order Month Name"]
    + " "
    + monthly_sales_display["Order Year"].astype(str)
)

monthly_sales_display["Sales"] = monthly_sales_display["Sales"].map(
    lambda x: f"${x:,.2f}"
)

monthly_sales_display = monthly_sales_display[["Month", "Sales"]]

monthly_sales_display
```

#### Monthly Sales Trend Visualization

The following line chart illustrates the monthly sales trend over the analysis period. It helps identify periods of sales growth, seasonal fluctuations, and overall business performance across different months.

```python
# Create monthly sales trend chart

plt.figure(figsize=(16, 8))

sns.lineplot(
    data=monthly_sales_display,
    x="Month",
    y=monthly_sales["Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales (USD)")

plt.xticks(rotation=90)

plt.ticklabel_format(style="plain", axis="y")

plt.tight_layout()

# plt.savefig(
#     "../outputs/figures/05_monthly_sales_trend.png",
#     dpi=300,
#     bbox_inches="tight"
# )

plt.show()
```

#### Key Business Insights

- Monthly sales show a clear upward trend from **2011 to 2014**, indicating continuous business growth over the four-year period.

- The **highest monthly sales** were recorded in **November 2014**, reaching **\$555,279.03**, followed by **December 2014** with **$503,143.69**.

- The **lowest monthly sales** occurred in **February 2011**, totaling **$91,152.16**, representing the weakest monthly performance in the dataset.

- Monthly sales exceeded **$400,000** several times during **2014**, demonstrating significantly stronger business performance compared to previous years.

- Sales generally remained below **$350,000** during **2011** and **2012**, while **2013** and especially **2014** experienced noticeable improvements in monthly revenue.

- The fluctuations observed throughout the months suggest seasonal demand patterns, while the overall increasing trend indicates sustained business expansion over time.

### Correlation Analysis

Correlation analysis measures the strength and direction of relationships between numerical business variables. It helps identify which features are positively or negatively associated, supporting better business understanding and decision-making.

```python
# Select numerical features

correlation_data = df_clean[
    ["Sales", "Profit", "Quantity", "Discount", "Shipping Cost"]
]

# Calculate correlation matrix

correlation_matrix = correlation_data.corr()

correlation_matrix
```

#### Correlation Heatmap

The following heatmap visualizes the correlation among the main numerical business features. Strong positive and negative correlations help identify relationships between sales, profit, quantity, discount, and shipping cost.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.tight_layout()

# plt.savefig(
#     "../outputs/figures/10_correlation_heatmap.png",
#     dpi=300,
#     bbox_inches="tight"
# )

plt.show()
```

#### Key Business Insights

- **Sales** and **Shipping Cost** have the strongest positive correlation (**0.77**), indicating that higher-value orders generally incur higher shipping costs.

- **Sales** and **Profit** show a **moderate positive correlation (0.48)**, suggesting that increased sales generally lead to higher profits, although other business factors also influence profitability.

- **Sales** and **Quantity** have a **weak positive correlation (0.31)**, indicating that higher sales are not always driven by purchasing larger quantities.

- **Discount** has a **moderate negative correlation with Profit (-0.32)**, showing that higher discounts tend to reduce profitability.

- **Discount** has a **very weak negative correlation with Sales (-0.09)** and **Quantity (-0.02)**, suggesting that offering larger discounts does not significantly increase sales volume or the number of items sold.

- Overall, the correlation analysis indicates that **Sales, Profit, and Shipping Cost** are positively related, while **Discount** negatively impacts profitability and has only a limited relationship with other business metrics.
