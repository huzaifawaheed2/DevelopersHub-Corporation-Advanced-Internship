# Interactive Business Dashboard in Streamlit

## Introduction

Business dashboards provide an interactive way to analyze sales performance and support data-driven decision-making. Instead of examining raw data, dashboards summarize key business metrics through interactive charts, filters, and Key Performance Indicators (KPIs).

In this project, an interactive business dashboard is developed using Streamlit and the Global Superstore dataset. The dashboard allows users to explore sales, profit, customer performance, and product categories through dynamic filters and visualizations. It enables business users to quickly identify trends, compare regional performance, and make informed decisions based on real-time insights.

## Problem Statement

Businesses generate large volumes of sales data every day. Analyzing this information using spreadsheets alone is time-consuming and makes it difficult to identify important trends and business opportunities.

To improve decision-making, an interactive dashboard is required that enables users to filter sales data by Region, Category, and Sub-Category while displaying important business metrics such as Total Sales, Total Profit, Top Customers, and sales performance through interactive visualizations.

## Project Objectives

The main objectives of this project are:

- Perform data cleaning and preprocessing.
- Prepare the dataset for dashboard development.
- Develop an interactive dashboard using Streamlit.
- Implement interactive filters for Region, Category, and Sub-Category.
- Display key performance indicators (KPIs) such as Total Sales, Total Profit, Total Orders, and Total Customers.
- Visualize business performance using interactive charts.
- Identify the Top 5 Customers by Sales.
- Support business decision-making through interactive data storytelling.

## Dataset Information

This project uses the **Global Superstore Dataset**, which is widely used for business intelligence, sales analytics, and dashboard development projects.

**Dataset Source:**  
https://www.kaggle.com/datasets/apoorvaappz/global-super-store-dataset

The dataset contains detailed sales transaction records from a global retail superstore, including customer information, product details, order history, shipping data, and financial metrics.

The primary objective of this dataset is to analyze business performance, identify sales trends, evaluate regional and product-wise performance, and support data-driven business decision-making.

Each record represents an individual sales transaction and includes information such as customer details, product category, region, market, sales, profit, quantity, discount, shipping cost, and order priority.

This dataset is well suited for developing **interactive business intelligence dashboards** using **Streamlit**, enabling users to explore sales performance through dynamic filters, key performance indicators (KPIs), and interactive visualizations.
